```python
from bs4 import BeautifulSoup 
import requests 
import csv


def getDate(post):
    
    #Date is present inside div tag of bd-item class.
    date = post.find('div',class_="bd-item")
    
    #filter span for date
    date = date.find('span')
    
    #returns only date by ignoring tags.
    return date.text
    
    

def getTitle(post):
    
    #filter post by div tag and content class
    article_content = post.find('div',class_='content')
    article_content = post.find('h6')
    
    #Title is present inside anchor tag.
    title = article_content.find('a')
    
    #returns only title name by ignoring tags.
    return title.text



def getImgURL(post):
    
    
    img_div = post.find('div',class_='img')
    img_a = img_div.find('a')

    #attrs returns the attributes of the bg-img
    attribute = img_a.attrs
    
    return attribute['data-bg']




def getLikes(post):
    
    
    #likes count is present inside the a tag of zilla-likes class.
    likes = post.find('a',class_='zilla-likes')
    likes = likes.find('span')
    
    #we need only number 
    likes_count = likes.text.split()
    
    
    return likes_count[0]
    
    
    
    

def getDetails(s):
    
    res = []
    for i in s:
        #dictonary to store the values
        d = {}
        
        d['Date'] = getDate(i)
        d['Title'] = getTitle(i)
        d['Img_url'] = getImgURL(i)
        d['Likes'] = getLikes(i)
        
        res.append(d)
    
    return res
        
        
        
    

def findArticles(soup):
    #find_all returns all the tags with name article and blog-item category-blog with-imag as class
    s = soup.find_all('article', class_='blog-item category-blog with-image')
    
    return s

    

#result store the Blog title,img url,date,likes count data of all pages.
result=[]

#there are 45 blog pages. So iterate through each page.
for i in range(1,46):
   
    #Blog page Url
    url = 'https://rategain.com/blog/'+"page/"+str(i)+"/"
   
    #To over come form 403 forbidden error
    headers = {
        'User-Agent':'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Mobile Safari/537.36'
    }

    response = requests.get(url, headers= headers)
    
    #parser the web page
    soup = BeautifulSoup(response.content, 'html.parser') 
    
    #s contains all articles in a page
    s = findArticles(soup)
    
    #resultOfPage stores all the Blog title,img url,date,likes count of an page.
    resultOfPage = getDetails(s)
    
    #extend the result list
    result.extend(resultOfPage)
    
    

    
#store the result in a excel file


filename = 'post_details.csv'
with open(filename, 'w', newline='',encoding="utf-8") as f: 
    w = csv.DictWriter(f,['Date','Title','Img_url','Likes']) 
    w.writeheader() 
      
    w.writerows(result)


    
    

```
